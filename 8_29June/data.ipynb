```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.datasets import make_regression
```

```python
# make_regression
# Generate a random regression problem.
# The output is generated by applying a (potentially biased) random linear regression model with n_informative
# nonzero regressors to the previously generated input and some gaussian centered noise with some adjustable scale.

X,y = make_regression(n_samples = 500,n_features=1,noise=1.91)
print(X.min(),X.max())
print(y.min(),y.max())
```

```python
print(X) # it's a 2d array
print(y) # 1d array
```

```python
plt.scatter(X[:, 0], y[:])
```
