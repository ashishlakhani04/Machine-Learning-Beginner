# Matplotlib

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#plot simple ranges
x = np.linspace(start=-5,stop=10,num=10)
y = 2*x

print(x)
print(y)
plt.figure(0)
plt.title('Line')
plt.grid(True)
plt.plot(x,y,'*-r')

plt.show()
```

```python
# Scatter plot of line
X = np.random.random((500,1))
Y = np.random.random((500,1))

plt.figure(0)
plt.plot(X,Y,'.')
```

```python
#plot sine and cosine
x = np.linspace(start=-2*np.pi,stop=2*np.pi,num=100)

plt.figure(0)
plt.plot(x, np.sin(x), '--')
plt.plot(x, np.cos(x), 'g-')
plt.show()
```

```python
# plot sine-wave,cosine-wave,and some lines with subplot
plt.figure(0)
plt.subplot(2, 2, 1)
plt.plot(x, np.cos(x), 'g-')

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), 'g-')

plt.subplot(2, 2, 3)
plt.plot(x, np.cos(x), 'g-')

plt.show()
```
